```python
class CLL:
    class _Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    class CircularLinkList:
        def __init__(self):
            self.head = None
            self.count = 0

        def insert(self ,data):
            new_node = CLL._Node(data)
            if self.head == None:
                self.head = new_node
                self.head.next = self.head
            else:
                temp = self.head
                while temp.next != self.head:
                    temp = temp.next
                temp.next = new_node
                new_node.next = self.head

        def search(self, data):
            if self.head.data == data:
                return True
            else:
                temp = self.head
                while True:
                    if temp.data == data:
                        return True
                    else:
                        temp = temp.next

                    if temp.next == self.head:
                        break
                return False    
            
        def count_data(self):
            self.count = 0
            temp = self.head

            while True:
                self.count += 1
                if temp.next == self.head:
                    break
                temp = temp.next
            return self.count
        
        def center(self):
            temp = self.head
            if self.count_data()%2 == 0 :
                for _ in range( 1 ,self.count_data()//2):
                    temp = temp.next
                return temp.data
            else:
                for _ in range(1 , round(self.count_data()/2+0.5)):
                    temp = temp.next
                return temp.data

        def delete(self,data):
            if self.head.data == data:
                temp_one = self.head
                temp_two = self.head
            
                while True:
                    if temp_two.next == self.head:
                        temp_three = self.head.next
                        temp_two.next = temp_three
                        self.head = temp_three
                        del temp_one
                        temp_one = None
                        break
                    temp_two = temp_two.next
                    
            else:
                temp_one = self.head
                
                while True:
                    if temp_one.next.data == data:
                        temp_two = temp_one.next
                        temp_one.next = temp_one.next.next
                        del temp_two
                        temp_two = None
                    
                    temp_one = temp_one.next

                    if temp_one.next == self.head:
                        break


        def display(self):
            temp = self.head

            while True:
                print(temp.data)
                temp = temp.next
                if temp == self.head:
                    break

obj = CLL()
circularLL = obj.CircularLinkList()
circularLL.insert(10)
circularLL.insert(20)
circularLL.insert(30)
circularLL.insert(40)
circularLL.insert(50)

print(" --- Display ---")
circularLL.display()

print(" --- Count ---")
print(circularLL.count_data())

print(" --- Center ---")
print(circularLL.center())

print(" --- Search ---")
print(circularLL.search(30))

print(" --- Delete ---")
print(circularLL.delete(50))

print(" --- Display ---")
circularLL.display()

print(" --- Center ---")
print(circularLL.center())

```

```python
class DLL:
    class _Node:
        def __init__(self, data):
            self.data =data
            self.next = None
            self.prev = None

    class DoubleLinkList:
        def __init__(self):
            self.head = None
            self.count = 0
        
        def insert(self, data):
            new_node = DLL._Node(data)

            if self.head == None:
                self.head = new_node
            else:
                temp_one = self.head
                temp_two = self.head

                while temp_one.next !=None:
                    temp_two = temp_one
                    temp_one = temp_one.next
                
                temp_one.next = new_node
                temp_one.prev = temp_two

        def delete(self , data):
            if self.head.data == data:
                temp = self.head
                self.head = self.head.next
                self.head.prev = None
                del temp
                temp = None
            else:
                temp_one = self.head
                while temp_one != None:
                    if temp_one.data == data:
                        if temp_one.next:
                            temp_one.next.prev = temp_one.prev
                        if temp_one.prev:
                            temp_one.prev.next = temp_one.next
                        del temp_one
                        return
                    temp_one = temp_one.next

        def search(self ,data):
            temp = self.head
            while temp != None:
                if temp.data == data:
                    return True
                temp = temp.next
            return False

        def count_data(self):
            self.count = 0
            temp = self.head
            while temp != None:
                self.count +=1
                temp = temp.next
            return self.count

        def display(self):
            temp = self.head
            while temp != None:
                print(temp.data)
                temp = temp.next

obj = DLL()
doubleLinkList = obj.DoubleLinkList()
doubleLinkList.insert(10)
doubleLinkList.insert(20)
doubleLinkList.insert(30)
doubleLinkList.insert(40)
doubleLinkList.insert(50)
doubleLinkList.insert(60)
doubleLinkList.insert(70)
doubleLinkList.insert(80)
doubleLinkList.insert(90)


print(" --- Display --- ")
doubleLinkList.display()

doubleLinkList.delete(60)

print(" --- Display --- ")
doubleLinkList.display()

print(" --- Search --- ")
print(doubleLinkList.search(40))

print(" --- Count --- ")
print(doubleLinkList.count_data())
```
